## Selecting and retrieving patient info based on the condition of the patient

```
df_15 = _deepnote_execute_sql('SELECT \n    CASE \n        WHEN PatientStatus = \'Recovering\' THEN CONCAT(\'Birthday of recovering patient is \', PatientBirthday)\n        WHEN PatientStatus = \'Critical\' THEN CONCAT(\'Birthday of critical patient is \', PatientBirthday)\n        ELSE CONCAT(PatientStatus, \' is not a patient\')\n    END AS Patient_Info\nFROM Patients\nWHERE PatientStatus IN (\'Recovering\', \'Critical\');', 'SQL_DEEPNOTE_DATAFRAME_SQL', audit_sql_comment='', sql_cache_mode='cache_disabled')
df_15
```

## Getting average salary

```
df_16 = _deepnote_execute_sql('SELECT AVG(MonthlySalary) as AvgSalary FROM Salary;', 'SQL_DEEPNOTE_DATAFRAME_SQL', audit_sql_comment='', sql_cache_mode='cache_disabled')
df_16
```
